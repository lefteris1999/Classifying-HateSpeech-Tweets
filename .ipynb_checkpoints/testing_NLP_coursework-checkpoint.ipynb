# Next Step is to remove unnecessary/unwanted information from the tweets.
Such information include html/xml tags, special characters and "@user" which is an encoded form of an original Twitter user being tagged on the tweet.

```python
unwanted_symbols = re.compile('[ð]')

def cleaning(tweet):

  #temp_tweets = hf_merged["tweet"] # temporary list of tweets
  temp_tweets = []

  #convert all tweets to lowercase to ensure continuity when handling tweets. Uppercase and Lowercase plays no role in our task. Hate speech is hate speech whether it is written in uppercase, lowercase or a mixture of both.
  tweet = tweet.lower() 

  #In the following line we remove all tags ("@") since this denote a user and we do not care to whom the tweet refers to. In addition, it could be considered personal data and thus we want to scrap them.
  #Consideration to remove hashtags was considered, but rejected, since hashtags could possibly include hate/offensive language and thus removing them would jeopardize our model.
  tweet = re.sub("@[A-Za-z0-9_]+","", tweet)

  #Similar to the above line, the following 2 lines remove links, (http or www format)
  tweet = re.sub(r"http\S+", "", tweet)
  tweet = re.sub(r"www.\S+", "", tweet)

  #In this line we remove all punctuation from the tweet and replace it with a whiteline. The whiteline ensures that after removal of punctuation, words are not concatanated as one when they should be separated.
  tweet = re.sub(r"[^\w\d'\s\-]", ' ', tweet)

  # remove some symbols found in our data
  tweet = unwanted_symbols.sub('', tweet) 

  #Remove numbers
  tweet = re.sub(r"[0-9]","", tweet)

  #Here we remove any potential emojis in the Tweet. This code was found at https://stackoverflow.com/questions/33404752/removing-emojis-from-a-string-in-python
  emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                          "]+", flags=re.UNICODE)
    
  tweet = emoji_pattern.sub(r'', tweet)

  #Remove any extra whitespaces that might have been generated from cleaning
  tweet = re.sub(r'\s+', ' ', tweet)

  # #Append "cleaned" tweet to our list
  # temp_tweets.append(tweet)

  return tweet
```

Next step is to tokenize our tweets. Tokenization is the process of splitting a sentence into individual words (tokens) that appear in it.

```python
# function to tokenize tweets (split the tweet into individual words that appear in it)
def tokenize(tweet):

  tokens_tweet = word_tokenize(tweet)
  #print(tokens_tweet)

  return tokens_tweet
```

Now let us remove the stopwords from our data. Stopwords are commonly used words in a language that bare little to no significance in a sentence as they are used very frequently regardless of context or domain. Stopwords are "words that occur commonly across all the documents in the corpus". The NLTK library provides a list of stopwords that we can use to filter them out from our dataset. However, we will be expanding this to include some twitter/online specific "slang" that may not be included in the nltk list. The extension list was obtained from personal experiance and online research e.g. https://www.webopedia.com/reference/text-abbreviations/

```python
# Removing Stopwords

stopwords = nltk.corpus.stopwords.words("english") #nltk already has a list of stopwords. We can use this as a base and built upon it.

twitter_stopwords = ["ff", "rt", "#ff", "rofl", "lol", "lmk", "ily", "yolo", "smh", "lmfao", "nvm", "ikr", "ofc", 'ð', '', "'re", "'s", "'m", ] # extra custom stopwords
stopwords.extend(twitter_stopwords)

def remove_stopwords(tokens):

  cleaned_tokens = [t for t in tokens if not t in stopwords]

  return cleaned_tokens
```

Following the removal of stopwords, the next step is to lemmatize our tokens. Lemmatization is the process of extracting the simplest/most basic form of a word while considering its context and meaning. This is different to Stemming, which strips the word of its last few characters regarding of the meaning of context of the word.

However, before being able to lemmatize our tokens, we need to determine their appropriate part of speech tag i.e., whether the token refers to a noun, adjective or verb. This plays an important role, as the lemmatizer will come up with different, appropriate lemmas based on the tag of the token. A common example is the token "better", if it is passed as a noun the lemma will remain "better", but if it is passed as an adjective, the lemma will be "good".

```python
# Function to get part of speech tag of the token

def my_token_tag(token):

  tagged_token = nltk.pos_tag(token)

  return tagged_token
```

```python
# Function to lemmatize

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def lemmatize_tokens(tagged_token):

  lemmas = []
  #print(tagged_token)
  #print(type(tagged_token))
  token_tag = [tup[1] for tup in tagged_token]
  token_word = [tup[0] for tup in tagged_token]

  #print(token_tag)
  #print(type(token_tag))

  for x in range(len(token_tag)):
    tag = token_tag[x]
    word = token_word[x]
    lemmatized_token = lemmatizer.lemmatize(word, pos = get_wordnet_pos(tag))
    lemmas.append(lemmatized_token)
  
  return lemmas 
```

Now we need to convert these tokens to an appropriate and understandable form to train our models. We can do this using a CountVectorizer or a TF-IDF Vectorizer. A CountVectorizer simply counts the number of occurances of a word(token) in a document - in our case, a tweet. The TF-IDF Vectorizer however, takes into consideration the importance of each word to the whole corpus (collection of tweets). The TF-IDF is generally regarded as a better method as it can suppress the impact of frequently used, unimportant tokens such as "the" or "and". These tokens may appear multiple times in a tweet and across various tweets, but bear little to no information about the content or context of the tweet.

```python
#Implement TF-IDF Vectorizer

#Firstly we need to combine the tokens of the individual tweets - have a list of tokenised tweets
#At the moment we have a list of lists of tokens

def combine_tokens(tokenized_tweet):
  # print(tokenized_tweet)
  # print(type(tokenized_tweet))
  # print("\n")
  combined_tokens = ""
  for token in tokenized_tweet:
    #tokenized_tweet = ' '.join(tokenized_tweet[i])
    #print(token)
    #print(type(token))
    combined_tokens = combined_tokens + ' ' + token
    #print(combined_tokens)
    #print("\n")
  return combined_tokens
```

```python
cleaned_tweet = []
tokens_clean = []
tokens_stop = []
tagged_tokens = []
lemmatized_tokens = []
combined_lemmas = []

#cleaning

for tweet in X:
  cleaned_tweet.append(cleaning(tweet))

for i in range(0, 3):
  print(cleaned_tweet[i])

# print("length of set:", len(cleaned_tweet))
# print(type(cleaned_tweet))

#tokenizing

for tweet in cleaned_tweet:
  tokens_clean.append(tokenize(tweet))

for i in range(0, 3):
  print(tokens_clean[i])

#Remove stopwords

for token in tokens_clean:
  tokens_stop.append(remove_stopwords(token))

for i in range(0, 3):
  print(tokens_stop[i])

#get part of speech tag

for token in tokens_stop:
  tagged_tokens.append(my_token_tag(token))

for i in range(0, 3):
  print(tagged_tokens[i])

#lemmatize

for token_tag in tagged_tokens:
  lemmatized_tokens.append(lemmatize_tokens(token_tag))

for i in range(0, 3):
  print(lemmatized_tokens[i])

#combine lemmatized tokens as a sentence so that we can run tf-idf later on

for lemma_token in lemmatized_tokens:
  combined_lemmas.append(combine_tokens(lemma_token))

for i in range(0, 3):
  print(combined_lemmas[i])
```

###Zero Rate Baseline:

First lets establish a baseline. This baseline will help us evaluate our models. Baseline models are simple and act as a benchmark accuracy that our more advanced models should at least try to match. One such baseline model is called the Zero Rate Classifier (ZeroR) which just predicts the most frequent label in the dataset.

Comparing our model to the baseline will help us evaluate and see whether our model truly makes informed classifications in comparison to simple algorithms that randomly predict or predict most frequent class.

```python
# zero rule algorithm for classification
def zero_ruleBase(label_train, x_test):
 zeroRule = hf_merged["label"].mode() # get the mode label of our dataset. The label that appears the most frequent
 predicted = [zeroRule for i in range(len(x_test))] # this "model" always predicts the label which appeard the most in the dataset
 return predicted

zeroR_baseline = zero_ruleBase(y, X_test)
print(len(zeroR_baseline))
print(type(zeroR_baseline))
np.asarray(zeroR_baseline).shape

baseline_accuracy = metrics.accuracy_score(y_test, zeroR_baseline) # accuracy of Zero-Rule prediction
print(baseline_accuracy)
```

```python
cleaned_tweet_lstm = []

#cleaning

for tweet in X_lstm:
  cleaned_tweet_lstm.append(cleaning(tweet))

for i in range(0, 3):
  print(cleaned_tweet_lstm[i])

hf_merged["tweet"] = cleaned_tweet_lstm

hf_merged.head()
# print("length of set:", len(cleaned_tweet_lstm))
# print(type(cleaned_tweet_lstm))
```

```python
MAX_WORDS = 50000 # The max number of words to be used - 50000
MAX_TWEET_LENGTH = 250 # Max number of words in each complaint - 250
EMBEDDING_DIM = 100 # dimension of embedding

tokenizer = Tokenizer(num_words=MAX_WORDS, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True) # specify number of tokens to be extracted from tokenizer (MAX_WORDS) and preprocessing techniques to be used(cleaning by removing symbols specified in the "filters" argument and turning tweet to lowercase)
tokenizer.fit_on_texts(hf_merged["tweet"].values) # tokenize our tweets
word_index = tokenizer.word_index # get the word indexes
print('%s Words.' % len(word_index)) # print number of tokens(words) extracted
```

```python
X = tokenizer.texts_to_sequences(hf_merged['tweet'].values)
X = pad_sequences(X, maxlen=MAX_TWEET_LENGTH) #pad sequences so that all sequences are of the same length
print(X.shape)
```

```python
print(X)
y_lstm = pd.get_dummies(hf_merged['label']).values # convert categorical variable "label" into idicator columns. since there are 3 distinct labels, there will be 3 columns in this dataframe indicating what label the tweet is. This need to be done in order to be able to later pass it to the lstm
print(y_lstm)
hf_merged.head()
```

```python
# Split into training/test sets
X_train_lstm, X_test_lstm, y_train_lstm, y_test_lstm = train_test_split(X, y_lstm, test_size=0.2, random_state=123, stratify = y)

print("x train:", len(X_train_lstm))
print("x test:", len(X_test_lstm))
print("y train:", len(y_train_lstm))
print("y tests:",len(y_test_lstm))
```

```python
model = Sequential() # specify sequential mode
model.add(Embedding(MAX_WORDS, EMBEDDING_DIM, input_length=X.shape[1])) #1st layer
model.add(tf.keras.layers.Dropout(0.5)) # Spatial dropout of 20% - 
model.add(tf.keras.layers.LSTM(100, dropout=0.5, recurrent_dropout=0.5)) # 2nd layer - LSTM
model.add(tf.keras.layers.Dense(3, activation='softmax')) # 3 outputs since we have 3 possible classes/labels and activation set to "softmax" - better for multiclass problems
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) # use categorical cross entropy since our problem is of multi-class nature - for binary classification, seen later, we can use binary cross entropy. Optimizer set to adam

batch_size = 64 # specify batch size
epochs = 2 # we will only train the model for 2 epochs. More epochs will take a lot of time and with two we already get a good accuracy

history = model.fit(X_train_lstm, y_train_lstm, epochs = epochs, batch_size = batch_size, validation_split = 0.1, callbacks = [tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)]) # fit the model and store metrics in variable
```

```python
MAX_WORDS = 50000
MAX_TWEET_LENGTH = 250
EMBEDDING_DIM = 100

tokenizer = Tokenizer(num_words=MAX_WORDS, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(hf_merged["tweet"].values)
word_index = tokenizer.word_index
print('%s tokens.' % len(word_index))
```

```python
X_bal_lstm = tokenizer.texts_to_sequences(hf_merged['tweet'].values)
X_bal_lstm = pad_sequences(X, maxlen=MAX_TWEET_LENGTH)
print(X_bal_lstm.shape)
```

```python
print(X)
y_lstm_bal = pd.get_dummies(hf_merged['label']).values
print(y_lstm_bal)
hf_merged.head()
```

```python
# Split into training/test sets
X_train_lstm_bal, X_test_lstm_bal, y_train_lstm_bal, y_test_lstm_bal = train_test_split(X_bal_lstm, y_lstm_bal, test_size=0.2, random_state=123, stratify = y)

print("x train:", len(X_train_lstm_bal))
print("x test:", len(X_test_lstm_bal))
print("y train:", len(y_train_lstm_bal))
print("y tests:",len(y_test_lstm_bal))
```

```python
lstm_bal = Sequential()
lstm_bal.add(Embedding(MAX_WORDS, EMBEDDING_DIM, input_length=X_bal_lstm.shape[1]))
lstm_bal.add(tf.keras.layers.Dropout(0.5))
lstm_bal.add(tf.keras.layers.LSTM(100, dropout=0.5, recurrent_dropout=0.5))
lstm_bal.add(tf.keras.layers.Dense(2, activation='softmax')) # this time we have 2 labels so 2 specify 2 outputs
lstm_bal.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # use binary cross entropy here as we reduced the problem to a binary classification

epochs = 2
batch_size = 64

history_bal = lstm_bal.fit(X_train_lstm_bal, y_train_lstm_bal, epochs=epochs, batch_size=batch_size,validation_split=0.1,callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)])
```
